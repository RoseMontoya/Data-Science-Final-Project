### Combining individual datasets together.

```python
master_mood = pd.concat([happy_vibes, chill_vibes, sad_vibes, romance_vibes, confidence_vibes])
master_mood
```

### Reset index

```python
master_mood1 = master_mood.reset_index()
master_mood1
```

### Removing columns that contain old indexes

```python
master_mood2 = master_mood1.drop(['Unnamed: 0', 'index'], axis = 1)
master_mood2
```

### Look at songs that are listed in multiple categories (assumption of independence)

```python
dup_songs = master_mood2.loc[master_mood1.duplicated(subset = ['Name'], keep = False), :]
```

```python
dup_songs.sort_values(by='Name')
```

### There were only 9 duplicate songs. I went through and listened to them, and I am going to keep them in the category I feel they fit best in based on lyrical contents. 

```python
master_no_dupes = master_mood2.drop([705, 830, 172, 12, 152,193, 77,7, 182])
master_no_dupes
```

### Recoding Mood Genres 

```python
def mood_recode(mood):
    if mood == 'Happy':
        return 1
    if mood == 'Chill':
        return 2
    if mood == 'Sad':
        return 3
    if mood == 'Romance':
        return 4
    if mood == 'Confidence':
        return 5
```

```python
master_no_dupes['MoodR']= master_no_dupes['Mood'].apply(mood_recode)
master_no_dupes
```

### Recode Explicit 

```python
def explicit_recode(song):
    if song == False:
        return 0
    if song == True:
        return 1
```

```python
master_no_dupes['ExplicitR']= master_no_dupes['Explicit'].apply(explicit_recode)
master_no_dupes
```
