## Correlation Matrix
### I wanted to look at a correlation matrix of my continious variables, both together as a whole and seperately in each mood genre to see if there was any difference. 

```python
mood_cont = mood_cleaned.drop (['Key', 'Mode', 'MoodR', 'ExplicitR'], axis=1)
```

```python
plt.figure(figsize=(16, 6))
# define the mask to set the values in the upper triangle to True
mask = np.triu(np.ones_like(mood_cont.corr(), dtype=bool))
heatmap = sns.heatmap(mood_cont.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='viridis')
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=16);
```

### The strongest correlation we have is between acousticness and energy. It's a negative correlation, so the more acoustic a song is then less energy it has. Next, we have an almost strong correlation between loudness and energy. And this correlation is positive, so the higher a loudness the more energy as well. There is a moderate correlation between acousticness and loudness. They have a negative correlation, so the more acoustic the less loud. Instrumentalness and popularity have a moderate negative correlation. So the more instrumentalness a song has the less popular it tends to be. Valence has a moderate, positive correlation with danceability, energy, and loudness. So the higher danceability, energy, and loudness a song had, the more valence it has. A few variables have a mild correlation, but there are so many variables that I'm not gonna delve into them. 

```python
happy = mood_cont.loc[mood_cleaned['Mood'] == 'Happy']

plt.figure(figsize=(16, 6))
mask = np.triu(np.ones_like(happy.corr(), dtype=bool))
heatmap = sns.heatmap(happy.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='viridis')
heatmap.set_title('Triangle Correlation Heatmap for Happy', fontdict={'fontsize':18}, pad=16);
```

### The Happy tracks have a much lower correlation between acousticness and energy than the combined chart.  In fact, Happy correlations are lower in all the significant correlations from above, except for danceability and duration with a -0.28 compared to the -0.27 correlation of the combined chart. So, Happy audio features tend to have a less significant correlations to each other. 

```python
chill = mood_cont.loc[mood_cleaned['Mood'] == 'Chill']

plt.figure(figsize=(16, 6))
mask = np.triu(np.ones_like(chill.corr(), dtype=bool))
heatmap = sns.heatmap(chill.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='viridis')
heatmap.set_title('Triangle Correlation Heatmap for Chill', fontdict={'fontsize':18}, pad=16);
```

### Most of the correlations are a bit lower, but there are some that really stand out as being much lower. When it came to the correlation between valence and energy/loudness, Chill had basically no correlation compared to the moderate correlation in all categories. This goes the same for speechiness and danceability. With all genres, there was a correlation of .34, and a correlation of .054 with only the chill category. 

```python
sad = mood_cont.loc[mood_cleaned['Mood'] == 'Sad']

plt.figure(figsize=(16, 6))
mask = np.triu(np.ones_like(sad.corr(), dtype=bool))
heatmap = sns.heatmap(sad.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='viridis')
heatmap.set_title('Triangle Correlation Heatmap for Sad', fontdict={'fontsize':18}, pad=16);
```

### Sad had the highest correlation between acousticness and energy with a -0.81. Overall, the sad category's correlations are more prominent, having higher correlations compared to the combined chart (loud-energy, acoustic-loud, instrumental-loud). But it does have basically no correlation between acousticness and danceability compared to the mild correlation seen in all the other charts. 

```python
romance = mood_cont.loc[mood_cleaned['Mood'] == 'Romance']

plt.figure(figsize=(16, 6))
mask = np.triu(np.ones_like(romance.corr(), dtype=bool))
heatmap = sns.heatmap(romance.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='viridis')
heatmap.set_title('Triangle Correlation Heatmap for Romance', fontdict={'fontsize':18}, pad=16);
```

### Romance has the strongest correlations in loudness-energy and acoustic-loudness then any of the other charts. Also the strongest correlation between valence and energy. 

```python
conf = mood_cont.loc[mood_cleaned['Mood'] == 'Confidence']

plt.figure(figsize=(16, 6))
mask = np.triu(np.ones_like(conf.corr(), dtype=bool))
heatmap = sns.heatmap(conf.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='viridis')
heatmap.set_title('Triangle Correlation Heatmap for Confidence', fontdict={'fontsize':18}, pad=16);
```

### Confidence has barely any correlation between acousticness and energy, a big difference from all the other graphs. Confidnece also has a nonexistent correlation between acousticness and loudness, another big difference from the rest of the graphs. Confidence did have a moderate correlation between speechiness-dance (positive) and danceability-duration (negative), a stronger correlation when compared to the other charts. 

### Key Significant Differences Overall:
### Confidence had much lower correlations in acousticness-energy and acousticness-loudness than the other charts. Only Combined, Happy, and Chill have a mild correlation between instrumentalness and popularity. Chill is the only category with no correlation between valence-engery. Chill and Happy have no correlation between valence-loudness, compared to the weak correlation in the others. Sad is the only category to have no correlation between acousticness-danceability, with other categories have a weak correlation. Happy and Confidence are the only categories with no correlation between instrumentalness-loudness, while the others have a weak correlation.
