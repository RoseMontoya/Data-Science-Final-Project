# Exploratory Analysis
## Load in Packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load in Dataframe

```python
master_mood = pd.read_csv('master_mood.csv')
master_mood
```

## Fix Index

```python
master_mood1 = master_mood.reset_index()
master_mood1
```

## Removing Unnecessary Columns

```python
master_mood2 = master_mood1.drop(['Unnamed: 0', 'index'], axis = 1)
master_mood2
```

```python
mood_cleaned = master_mood2.drop(['Name', 'Track_ID', 'Album', 'Artist','Artist_ID','Artist_Genre', 'Explicit', 'Playlist'], axis = 1)
mood_cleaned
```

## Removing Outliers
### Looking at overall distribution in features and identifying outliers.

```python
mood_cont = mood_cleaned.drop (['Key', 'Mode', 'MoodR', 'ExplicitR'], axis=1)
```

```python
flierprops = dict(marker='o', markersize=5, markeredgecolor='black', markerfacecolor='green', alpha=0.5)
p = sns.boxplot(x= mood_cont['Duration_ms'], palette = "viridis", flierprops=flierprops)
p.set_xlabel('Duration_ms', fontsize= 14, fontweight='bold')
p.set_title('Duration_ms Distribution', fontsize= 16, fontweight='bold');
```

```python
from numpy import percentile

q25, q75 = percentile(mood_cont['Duration_ms'], 25), percentile(mood_cont['Duration_ms'], 75)
iqr = q75 - q25

cut_off = iqr*1.5
lower, upper = q25-cut_off, q75 + cut_off

outliers= [x for x in mood_cont['Duration_ms'] if x < lower or x > upper]
len(outliers)
```

```python
duration_outliers = master_mood2.loc[mood_cont['Duration_ms'].isin(outliers)]
duration_outliers
```

### Combining outliers

```python
all_out = pd.concat([duration_outliers, pop_outliers, dance_outliers, loud_outliers, speech_outliers, instr_outliers, live_outliers, tempo_outliers])
all_out
```

```python
all_out2 = all_out.drop_duplicates(subset=['Track_ID'], keep='first')
all_out2
```

```python
df_no_out = master_mood2.loc[-master_mood2['Track_ID'].isin(all_out2['Track_ID'])]
df_no_out
```

```python
print(len(df_no_out.loc[df_no_out['Mood'] == "Happy"]))
print(len(df_no_out.loc[df_no_out['Mood'] == "Chill"]))
print(len(df_no_out.loc[df_no_out['Mood'] == "Sad"]))
print(len(df_no_out.loc[df_no_out['Mood'] == "Romance"]))
print(len(df_no_out.loc[df_no_out['Mood'] == "Confidence"]))
```
